#if all values are null,remove that row

```python
df.dropna(how='all')
```

#any column with nan..remove that whole row


```python
df.dropna(how='any')
```

#filling null values with 0




```python
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({'Physics':'none','chemistry':0})
```

#fill with avg

```python
df['Physics'].fillna(value=df['Physics'].mean())
```

#FORWARD FILL : fill the null value with its previous value
# BACKWARD FILL : fill with next value


```python
df.fillna(method='ffill')
```

#

```python
df.fillna(method='bfill')
```

```python
df.replace(to_replace=26,value=1000) #replace 26 with 1000
```

```python
df.replace(to_replace=[50,51,52,53,54,55,56],value='A')
```
